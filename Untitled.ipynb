```python
import pandas as pd

#dev = pd.read_xml(r'assin2-dev.xml')
#test = pd.read_xml(r'assin2-test.xml')
train = pd.read_xml(r'assin2-train-only.xml')

display(train)
```

```python
import spacy
nlp = spacy.load("pt_core_news_sm")    
```

```python

for i in range(len(train)):
    coluna_t = train.iloc[i,3]
    coluna_t = nlp(coluna_t)
    
    coluna_h = train.iloc[i,4]
    coluna_h = nlp(coluna_h)
    
    similaridade = coluna_t.similarity(coluna_h)
    print(f'{coluna_t} x {coluna_h} = {similaridade:.2f} \n') 
    
```
